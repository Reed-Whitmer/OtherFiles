```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

%matplotlib inline
```

```python
df = pd.read_csv('C:/Users/reedw\\CTBA\\CTBA\\M3\\ConstructionTimeSeriesDataV2.csv', header = 0)
print(df)
```

```python
dfYear = df.groupby(df['Month-Year'].astype(str).str[4:]).sum().reset_index()

dfYear = dfYear.drop(columns = ['Month'])
dfYear = dfYear.rename(columns={"Month-Year": "Year"})
dfYear = dfYear.drop([12])
print(dfYear)
```

```python
fig, ax = plt.subplots()
tick_spacing = 100000


plt.suptitle('Private vs Public Construction Contracts per Year', fontsize = 16)


plt.plot(dfYear['Year'], dfYear['Private Construction'], marker = 'o', label = 'Private Construction', color = 'k')
plt.plot(dfYear['Year'], dfYear['Public Construction'], marker = 'o', label = 'Public Construction', color = 'b')
plt.ylim(0,1000000)
ax.yaxis.set_major_locator(plt.MultipleLocator(tick_spacing))

plt.xlabel('Year', fontsize = 12)
plt.ylabel('Contract Values', fontsize = 12)
plt.grid(True)


plt.legend()
fig.set_size_inches(10,5)


```
